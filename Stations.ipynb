```python
import pandas as pd
import re
```

```python
stn = pd.read_csv('../ghcnd-stations.csv')
```

```python
stn['COUNTRY_CODE'] = stn['STATION'].str[0:2]
stn['STATE_CODE'] = stn['NAME'].str[0:2]
stn = stn.loc[stn['COUNTRY_CODE']=='US']
stn.head(15)
```

```python
stn['COUNTRY_CODE'].unique()
```

```python
def find_digit_pos(name):
    try:
        position = re.search(r"\d",name)
        return name[0:position.start()]
    except:
        return name[0:len(name)]
```

```python
# stn.loc[stn['NAME']=='NE HASTINGS 5.3 NNW'z
```

```python
stn["STATE_CITY_NAME"] = stn.apply(lambda row: find_digit_pos(row['NAME']),axis=1)
```
